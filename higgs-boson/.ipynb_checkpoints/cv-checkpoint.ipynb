## 数据探索
训练数据包含了 250000 事件，第一列是是事件id，30个特征列，一个权重列，以及一个标签列。

测试数据包含了 550000 个事件，30个特征值

所有变量都是 floating point 类型，除了 PRI_jet_num 为 integer。 以 PRI （PRImitives) 为的前缀特征为检测器测量得到的关于 bunch collision“原始” 数据. 以 DER （ DERived) 为 ATLAS 的物理学家选择的根据原始特征计算得到的数据。 缺失数据记为 −999.0, 与所有特征的正常值不同。

特征、权重和标签的具体语意可以查看 [CERN 的技术文档](http://higgsml.lal.in2p3.fr/documentation)。

```python
import numpy as np
import pandas as pd
import xgboost as xgb
```

```python
data_path = "./data"
dtrain = pd.read_csv(data_path + "/training.csv")
dtest = pd.read_csv(data_path + "/test.csv")
```

```python
dtrain.head()
```

```python
dtest.head()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
dtrain['Label'] = labelencoder.fit_transform(dtrain['Label'])
```

```python
# s => 1, b=> 0
dtrain.head()
```

计算正负比例的权重，在计算之前，先根据训练集和测试集的大小，调整下权重比值

```python
weight = dtrain.iloc[:,31] * len(dtest) / len(dtrain)
sum_pos = weight[dtrain['Label'] == 1].sum()
sum_neg = weight[dtrain['Label'] == 0].sum()
print ('weight statistics: wpos=%g, wneg=%g, ratio=%g' % ( sum_pos, sum_neg, sum_neg/sum_pos ))
```

```python
label  = dtrain.iloc[:,32]
data   = dtrain.iloc[:,1:31]
```

## xgboost常用参数介绍



```python
train = xgb.DMatrix(data, label, missing=-999.0,weight=weight)
```

```python
# setup parameters for xgboost
param = {}
# use logistic regression loss, use raw prediction before logistic transformation
# since we only need the rank
param['objective'] = 'binary:logitraw'
param['eta'] = 0.1
param['max_depth'] = 6
param['silent'] = 1
```

```python
def fpreproc(dtrain, dtest, param):
    label = dtrain.get_label()
    ratio = float(np.sum(label == 0)) / np.sum(label==1)
    # if (info.labels[i] == 1.0f) w *= param_.scale_pos_weight 
    param['scale_pos_weight'] = ratio
    wtrain = dtrain.get_weight()
    wtest = dtest.get_weight()
    sum_weight = sum(wtrain) + sum(wtest)
    wtrain *= sum_weight / sum(wtrain)
    wtest *= sum_weight / sum(wtest)
    dtrain.set_weight(wtrain)
    dtest.set_weight(wtest)
    return (dtrain, dtest, param)
```

```python
# boost 1000 trees
num_round = 1000
cvresult = xgb.cv(param, train, num_round, nfold=5,
       metrics={'ams@0.15', 'auc'}, early_stopping_rounds=10, seed = 0, fpreproc = fpreproc)
```

```python
cvresult
```

```python
n_estimators = cvresult.shape[0]

from matplotlib import pyplot
# plot
test_means = cvresult['test-ams@0.15-mean']
test_stds = cvresult['test-ams@0.15-std'] 
        
train_means = cvresult['train-ams@0.15-mean']
train_stds = cvresult['train-ams@0.15-std'] 

x_axis = range(0, n_estimators)
pyplot.errorbar(x_axis, test_means, yerr=test_stds ,label='Test')
pyplot.errorbar(x_axis, train_means, yerr=train_stds ,label='Train')
pyplot.title("HiggsBoson n_estimators vs ams@0.15")
pyplot.xlabel( 'n_estimators' )
pyplot.ylabel( 'ams@0.15' )
pyplot.savefig( 'HiggsBoson_estimators.png' )
pyplot.show()
```

```python
print ('train model using the best parameters by cv ... ')
bst = xgb.train( param, train, n_estimators );
```

```python
# save out model
bst.save_model('higgs_cv.model')

print ('train finished')
```
